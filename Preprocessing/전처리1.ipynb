```python
#@title 기본 제목 텍스트
#file to dataframe
#업로드한 파일을 DataFrame 형태로 변환

#import the library
import pandas as pd

#아래 코드는 각자가 업로드한 파일 이름으로 수정하셔서 사용하시면 됩니다.
filename = 'data2022_1.csv'

df = pd.read_csv(filename)
#'doc_new' 컬럼에 null 값을 공백으로 대체
df['doc_new'] = df['doc_new'].fillna('')

#data 미리보기
df.head()
```

### **형태소 분석 후 파일 저장**
* 형태소 분석기를 돌린 후 명사토큰을 'nouns' 열에 입력하고, 파일을 저장합니다.
* '.csv'형태로 저장되므로, 다음 전처리 과정인 '합성명사 사전 작성'을 위해서는 다시 구글 스프레드시트로 열어야 합니다.

```python
#import the library
#코드 실행에 필요한 라이브러리를 import하는 부분입니다.
from konlpy.tag import Mecab
mecab = Mecab()

noun = []

for element in df['doc_new']:
  noun.append(mecab.nouns(element))
df['nouns'] = noun

df.to_csv(filename, index=False)
```
